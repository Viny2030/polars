Una vez montado Google Drive, puedes acceder a los archivos dentro de tu unidad utilizando rutas est√°ndar de Python. Por ejemplo, si tienes un archivo CSV en tu Google Drive, puedes cargarlo en un DataFrame de pandas de la siguiente manera:<code>/content/drive/My Drive/tu_archivo.csv</code>

Si prefieres cargar el archivo directamente, tambi√©n puedes usar <code>gdown</code>. Esta es una herramienta √∫til para descargar archivos directamente desde Google Drive, utilizando solo el ID del archivo:
<BR><BR>
`https://drive.google.com/file/d/1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf/view?usp=sharing`
<BR><BR>
El ID del archivo es la parte despu√©s de `/d/` y antes de `/view?`. En este caso, ser√≠a `1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf`.
<BR><BR>
Luego se debe concatener la URL base de descarga (`https://drive.google.com/uc?id=`) de Google Drive con el ID del archivo. La estructura de la URL que debes construir es:
<BR><BR>
`https://drive.google.com/uc?id=1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf`

```python
ID = "1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf" # <-- debe indicar su ID
```

```python
import gdown

# Your file ID
ID = "1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf"

# URL base de descarga concatenado con el ID del archivo
url = f"https://drive.google.com/uc?id={ID}"

# Descargar el archivo
# quiet=True, desactiva la salida en pantalla durante la descarga (por defecto es False).
gdown.download(url, 'archivo.csv', quiet=True)

# Cargar el archivo descargado en un DataFrame
import pandas as pd # Import the pandas library
df = pd.read_csv('archivo.csv')
df.head()
```

## **<font color="DarkBlue">Repositorios</font>**

<p align="justify">
Es posible ingerir datos desde un repositorio web como el <a href = "https://archive.ics.uci.edu/">Repositorio de Aprendizaje Autom√°tico de UCI</a>, que contiene una amplia colecci√≥n de conjuntos de datos utilizados para investigaciones y educaci√≥n en Machine Learning. El siguiente c√≥digo lo encontrar√° en este <a href="https://archive.ics.uci.edu/dataset/2/adult">link</a>.

```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

# Obt√©n el dataset con id=2 (Adult dataset)
adult = fetch_ucirepo(id=2)

# Los datos (caracter√≠sticas) ya est√°n en un DataFrame de pandas
X = adult.data.features
y = adult.data.targets

# Combina las variables explicativas y la objetivo en un solo DataFrame
df = pd.concat([X, y], axis=1)
df.head()
```

<p align="justify">
Tambi√©n es posbile cargar el conjunto de datos usando <code>fetch_openml</code> de <code>scikit-learn</code>.
<br><br>
<a href="https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_openml.html">Documentaci√≥n</a>

```python
from sklearn.datasets import fetch_openml

# Cargar el conjunto de datos Adult Census desde sklearn
adult_census = fetch_openml(name="adult", version=2, as_frame=True)

# El conjunto de datos se carga como un Bunch, de donde podemos extraer el DataFrame
df = adult_census.frame
df.head()
```

```python
# Cargar el conjunto de datos Adult Census desde sklearn versi√≥n 3
adult_census = fetch_openml(name="adult", version=3, as_frame=True)

# El conjunto de datos se carga como un Bunch, de donde podemos extraer el DataFrame
df = adult_census.frame
df.head()
```

## **<font color="DarkBlue">Unidad local</font>**

Para cargar datos desde tu m√°quina local en Google Colab y trabajar con ellos sin necesidad de almacenar los datos en un repositorio en l√≠nea.

```python
from google.colab import files

# Subir el archivo
uploaded = files.upload()
```

Esto abrir√° un cuadro de di√°logo que te permitir√° seleccionar y subir archivos desde tu undidad local. Despu√©s de subir el archivo, se almacenar√° en el entorno de Colab. Puedes usar pandas para leer archivos CSV:

```python
# Hemos subido un archivo llamado 'adult_census.csv'
filename = list(uploaded.keys())[0]  # Obt√©n el nombre del archivo subido
df = pd.read_csv(filename)
```
